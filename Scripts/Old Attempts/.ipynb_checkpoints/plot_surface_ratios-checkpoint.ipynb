```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
from matplotlib import pyplot as plt
```

```python
# Load csv with surface area ratios
df = pd.read_csv('csv_data/all_uniprot_pdb_surface.txt', sep = ' ')
# Remove human proteins
df = df[~df['Organism'].isin(['human'])]
```

```python
df.rename(columns = {"surface_ratio": "Surface area ratio"}, inplace=True)
```

```python
df['Type'] = 'plant'
```

```python
df['Type'] = np.where(df['Organism'].astype(str).str.contains('whey'), 'Animal', 'Plant')
```

```python
df
```

```python
sns.displot(data=df, x='Surface area ratio', hue='Type', kind='kde', fill=True, palette=sns.color_palette('bright')[:2], height=5, aspect=1.5)
```
